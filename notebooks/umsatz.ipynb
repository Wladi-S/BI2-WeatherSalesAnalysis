```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('../data/raw/Umsatzdaten/Gastronomieumsätze_flat.csv', sep=';', encoding='ISO-8859-1')
```

```python
data['4_Auspraegung_Code'] = data['4_Auspraegung_Code'].str[-2:].astype(int)
data.rename(columns={'UMS002__Umsatz__2015=100': 'Umsatz'}, inplace=True)

data['Umsatz'] = data['Umsatz'].replace('...', float('NaN'))
data['Umsatz'] = data['Umsatz'].str.replace(',', '.').astype(float)
```

```python
data
```

```python
data = data[data.columns[data.nunique() > 1]]
```

```python
data['Zeit'] = pd.to_datetime(data['Zeit'].astype(str) + '-' + data['4_Auspraegung_Code'].astype(str), format='%Y-%m')
```

```python
data.sort_values(['5_Auspraegung_Label', 'Zeit'], inplace=True)
```

```python
px.line(data, y='Umsatz', x='Zeit', color='5_Auspraegung_Label')
```
